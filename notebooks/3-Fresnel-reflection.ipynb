```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual, FloatSlider
import ipywidgets as widgets 
%matplotlib inline
plt.style.use('seaborn')
```

```python
deg = 180/np.pi

def figure_plot(n_i,n_t):
    if n_i/n_t >=1:
        crit_angle = np.arcsin(n_t/n_i)
        x_i = np.linspace(0.0001,crit_angle,100)              # Incident Angles 
        x_t = np.arcsin((n_i/n_t)* np.sin(x_i))               # Transmitted Angles
    else:
        x_i = np.linspace(0.0001,np.pi/2,100)              # Incident Angles
        x_t = np.arcsin((n_i/n_t)* np.sin(x_i))                   # Transmitted Angles
    y = np.zeros(90)
    r_p = ((n_i*np.cos(x_i) - n_t*np.cos(x_t))/ (n_i*np.cos(x_i) + n_t*np.cos(x_t)))
    r_s = ((n_i*np.cos(x_t) - n_t*np.cos(x_i))/ (n_i*np.cos(x_t) + n_t*np.cos(x_i)))
    r_p = r_p**2
    r_s = r_s**2
    t_p = 1 - r_p
    t_s = 1 - r_s
    plt.figure(figsize=(15, 10))
    t_sp = plt.plot(x_i*deg,t_s, 'r', label='Transmission s-pol')
    r_sp = plt.plot(x_i*deg,r_s, 'r--', label='Reflection s-pol')
    t_pp = plt.plot(x_i*deg, t_p, 'g', label='Transmission p-pol')
    r_pp = plt.plot(x_i*deg, r_p, 'g--', label='Reflection p-pol')
    plt.title("Fresnel Equations")
    plt.xlim(0,90)
    plt.yticks(ticks=(np.linspace(0,1,11)))
    plt.ylim(-0.01,1.01)
    plt.legend()
    plt.show()
```

```python
n_i_widget = FloatSlider(
    min=1,
    max=2,
    step=0.01,
    value=1,
    continuous_update=True,
    readout_format='.2f')
n_t_widget = FloatSlider(
    min=1,
    max=2,
    step=0.01,
    value=1.2,
    continuous_update=True,
    readout_format='.2f')
```

```python
interactive_plot = interact(figure_plot, n_i=n_i_widget, n_t=n_t_widget);
```
