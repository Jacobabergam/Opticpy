```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
mpl.rcParams['figure.figsize'] = (20.0,10.0)
mpl.rcParams['image.cmap'] = 'viridis'
cmap = cm.viridis
```

```python
n=2**10
M = np.zeros((n,n))
I = np.arange(0,n)
x = I-n/2
print(x)
y = (n/2) - I
[X,Y] = np.meshgrid(x,y)
R = 10
```

```python
A = (X**2+Y**2 <= R**2)
```

```python
M[A] = 1
```

```python
plt.imshow(M, cmap=cmap)
plt.show()
DP = np.fft.fftshift(np.fft.fft2(M))
plt.imshow(abs(DP), cmap = cmap)
plt.show()
plt.imshow(abs(np.log2(DP)), cmap=cmap)
plt.show()
```

```python
n=2**10
M = np.zeros((n,n))
I = np.arange(0,n)
[X,Y] = np.meshgrid(x,y)
slitwidth = 15.0
slitheight = 1000.0
slitsepar = 50

x = I-n/2 + slitwidth/2 + slitsepar/2
y = (n/2) - I 
[X,Y] = np.meshgrid(x,y)
A = ((X < slitwidth/2) & (X > -slitwidth/2) & (Y < slitheight/2) & (Y> -slitheight/2))
M[A] = 1
x = I-n/2 - slitwidth/2 + -slitsepar/2
y = (n/2) - I
[X,Y] = np.meshgrid(x,y)
# A = (X**2+Y**2 <= R**2)
A = ((X < slitwidth/2) & (X > -slitwidth/2) & (Y < slitheight/2) & (Y> -slitheight/2))
M[A] = 0.1
```

```python
plt.imshow(M)
plt.show()
DP = np.fft.fftshift(np.fft.fft2(M))
plt.imshow(abs(DP))
plt.show()
plt.imshow(abs(np.log2(DP)))
plt.show()
```
