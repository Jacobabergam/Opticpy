# FDTD Algorythms

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML, Image
%matplotlib inline
```

## 1D FTDT evaluatoin

```python
dx = 0.1                # Space increment
dt = 0.05               # time increment
tmin = 0.0              # initial Time
tmax = 100.0              # simulate until
xmin = -5.0             # left bound
xmax = 5.0              # right bound
c = 2.0                 # Speed of sound
rsq = (c*dt/dx)**2      # finite diff solution
```

```python
nx = int((xmax-xmin)/dx) + 1    #number of points on the x grid
nt = int((tmax-tmin)/dt) + 2    #number of points on the t grid
u = np.zeros((nt,nx))           #solution to WE
```

```python
#set initial pulse shape
def init_fn(x):
    val = np.exp(-(x**2)/0.25)
    if val<0.001:
        return 0.0
    else:
        return val
```

```python
for a in range(0,nx):
    u[0,a]=init_fn(xmin+a*dx)
    u[1,a]=u[0,a]

#simulate dynamics
for t in range(1,nt-1):
    for a in range(1,nx-1):
        u[t+1,a] = 2 *(1-rsq)*u[t,a]-u[t-1,a]+rsq*(u[t,a-1]+u[t,a+1])
```

```python
fig = plt.figure()
plts = [] # get ready to populate this list the Line artists to be plotted
for i in range(nt):
    p, = plt.plot(u[i,:], 'k', linewidth=1) # this is how you'd plot a single Line
    plts.append( [p] )
ani = animation.ArtistAnimation(fig, plts,interval=50) # run the animation
HTML(ani.to_html5_video())
```

## Yee algorithm

```python
nx = 2**8
ny = 2**8
nt = 2**4

ex = np.zeros((nx,ny+1))
print(ex.shape)
ey = np.zeros((nx+1,ny))
print(ey.shape)
hz = np.zeros((nx,ny))
print(hz.shape)
cex = 8.85418782 * 10**-12
cey = 8.85418782 * 10**-12
chy = 4* np.pi * 10**-7
chx = 4* np.pi * 10**-7
```

```python
source = np.zeros(nt)
source[1:2]=1
hzx = int(nx/2)
hzy = int(ny/2)

# Source 
# e(x,y) = sin(n/5)
```

```python
ims=[]
for t in range(nt):
    plt.figure(figsize=(10,10),dpi=50)
    for i in range(nx):
        for j in range(1,ny-1):
            ex[i,j] = ex[i,j] + cex * (hz[i,j]-hz[i,j-1])
            
    for i in range(1,nx-1):
        for j in range(ny):
            ex[i,j] = ey[i,j] - cey * (hz[i,j] - hz[i-1,j])
    
    for i in range(nx):
        for j in range(ny):
            hz[i,j] = hz[i,j] + chy*(ex[i,j+1] - ex[i,j])- chx*(ey[i+1,j] - ey[i,j])
    hz[hzx:hzx+10,hzy:hzy+10] = source[t]
    plt.imshow(hz[hzx-10:hzx+20,hzy-10:hzy+20], animated=True)

ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)
plt.show()
```
