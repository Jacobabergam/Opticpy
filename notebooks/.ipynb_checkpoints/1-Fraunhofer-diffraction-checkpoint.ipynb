```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
mpl.rcParams['image.cmap'] = 'viridis'
cmap = cm.viridis
```

# Far Field Diffraction using FFT2 with Numpy

```python
n=2**10
M = np.zeros((n,n))
I = np.arange(0,n)
x = I-n/2
print(x)
y = (n/2) - I
[X,Y] = np.meshgrid(x,y)
R = 10
```

```python
A = (X**2+Y**2 <= R**2)
```

```python
M[A] = 1
```

```python
fig, axs = plt.subplots(1,3,figsize=(15,5))
axs[0].imshow(M, cmap=cmap)
DP = np.fft.fftshift(np.fft.fft2(M))
axs[1].imshow(abs(DP), cmap = cmap)
axs[2].imshow(abs(np.log2(DP)), cmap=cmap)
plt.show()
```

## Recreating the Double Slit

```python
n=2**10
M = np.zeros((n,n))
I = np.arange(0,n)
[X,Y] = np.meshgrid(x,y)
slitwidth = 10.0
slitheight = 100.0
slitsepar = 60

x = I-n/2 + slitwidth/2 + slitsepar/2
y = (n/2) - I 
[X,Y] = np.meshgrid(x,y)
A = ((X < slitwidth/2) & (X > -slitwidth/2) & (Y < slitheight/2) & (Y> -slitheight/2))
M[A] = 1
x = I-n/2 - slitwidth/2 + -slitsepar/2
y = (n/2) - I
[X,Y] = np.meshgrid(x,y)
# A = (X**2+Y**2 <= R**2)
A = ((X < slitwidth/2) & (X > -slitwidth/2) & (Y < slitheight/2) & (Y> -slitheight/2))
M[A] = 1
```

```python
fig, axs = plt.subplots(2,figsize=(15,10))
axs[0].imshow(M, cmap=cmap)
DP = np.fft.fftshift(np.fft.fft2(M))
axs[1].imshow(abs(DP[500:525,400:620]), cmap = cmap)
#axs[2].imshow(abs(np.log2(DP)), cmap=cmap)
plt.show()
```

## Double slit with variable transmission for one slit

```python
n=2**10
M = np.zeros((n,n))
I = np.arange(0,n)
[X,Y] = np.meshgrid(x,y)
slitwidth = 20.0
slitheight = 100.0
slitsepar = 90

x = I-n/2 + slitwidth/2 + slitsepar/2
y = (n/2) - I 
[X,Y] = np.meshgrid(x,y)
A = ((X < slitwidth/2) & (X > -slitwidth/2) & (Y < slitheight/2) & (Y> -slitheight/2))
M[A] = 1
x = I-n/2 - slitwidth/2 + -slitsepar/2
y = (n/2) - I
[X,Y] = np.meshgrid(x,y)
# A = (X**2+Y**2 <= R**2)
A = ((X < slitwidth/2) & (X > -slitwidth/2) & (Y < slitheight/2) & (Y> -slitheight/2))
M[A] = 0.1

fig, ax = plt.subplots(2,3, figsize=(15,10), dpi= 100)
fig.tight_layout()
for i in range(0,3):
    M[A] = 5/(i+0.000001)
    DP = np.fft.fftshift(np.fft.fft2(M))
    ax[0,i].imshow(M)
#     ax[1,i].imshow(abs(DP))
    ax[1,i].imshow(abs(np.log2(DP)))

fig.tight_layout()
plt.show()
```
